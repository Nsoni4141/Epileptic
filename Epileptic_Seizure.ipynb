```python
import pandas as pd
from Outlier_Detection_1 import find_outlier
import matplotlib.pyplot as plt 
from Outlier_detector_2 import detect_outlier
```

```python
data = pd.read_csv(r"Epileptic Seizure Recognition.csv")
data.isnull().sum()
data.head()
```

```python
x = data.loc[:,"X1":"X178"]
y = data['y']

detect_outlier( x )
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_transform = sc.fit_transform(x)
from sklearn.feature_selection import RFE
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection  import train_test_split  
model =RandomForestClassifier(n_estimators = 200 ,random_state=0)
x_train,x_test,y_train,y_acc = train_test_split(x_transform,y, test_size = 0.2, random_state = 0 )
```

```python
sel = RFE(model)
sel.fit(x_train,y_train) 
```

```python
transform_x_train = sel.transform(x_train)
transform_x_test = sel.transform(x_test)
```

```python
model.fit(transform_x_train,y_train)
```

```python
acc = model.score(transform_x_test,y_acc)
acc
y_pred = model.predict( transform_x_test)

```

```python
from sklearn.neighbors import KNeighborsClassifier 
model2 = KNeighborsClassifier(n_neighbors = 4)

model2.fit(transform_x_train,y_train)
acc2 = model2.score(transform_x_test,y_acc)

acc2
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred,y_acc)
```
